```python
import numpy as np 
import pandas as pd 
import random
from keras.datasets import cifar10

# import libraries for EDA and preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import np_utils

# Modeling
from keras.models import *
from keras.layers import *
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import plot_model
# Evaluating
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# create a list of class
classes = ['Airplane','Automobile','Bird','Cat','Deer','Dog','Frog','Horse','Ship','Truck']
```

```python
# visualize some images in train data
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i])
    ax.set(xticks=[], yticks=[], xlabel = classes[y_train[i][0]])
fig.suptitle('\nFigure 5. Train image Visualizations') 
plt.show() 
```

```python
# perform the one-hot encoding for labels
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
print("There are {} training images, labeled over {} categories.".format(y_train.shape[0], y_train.shape[1]))
print("There are {} test images, labeled over {} categories.".format(y_test.shape[0], y_test.shape[1]))
```

```python
# view some first rows of y_train after encoding
y_train[:5]
```

```python
# set random state
RANDOM_STATE = 42

# set batch size
BATCH_SIZE = 64

# set input shape
INPUT_SHAPE = x_train.shape[1:]

# set the number of categories
NUM_LABELS = y_train.shape[1]

# set the number of epochs
EPOCHS = 20

# set step_size_train and step_size_test
STEP_SIZE_TRAIN = x_train.shape[0]//BATCH_SIZE
STEP_SIZE_TEST = x_test.shape[0]//BATCH_SIZE
# train data augmentation
```

```python
# train data augmentation
train_datagen = ImageDataGenerator(featurewise_center=False,  # set input mean to 0 over the dataset  
                              featurewise_std_normalization=False,# divide inputs by std of the dataset
                              zoom_range = 0.2, # Randomly zoom image 
                              rotation_range = 20,  # randomly rotate images in the range (degrees, 0 to 180)
                              width_shift_range=3/32,  # randomly shift images horizontally (fraction of total width)
                              height_shift_range=3/32,  # randomly shift images vertically (fraction of total height)
                              horizontal_flip = True,  # randomly flip image
                              rescale = 1./255)    # multiply the data by the value provided
train_generator = train_datagen.flow(x_train, y_train, batch_size=BATCH_SIZE)

# test data augmentation
test_datagen = ImageDataGenerator(rescale = 1./255)
test_generator = test_datagen.flow(x_test, y_test, batch_size=BATCH_SIZE)
```

```python
# build CNN model
model1 = Sequential()
# first convolutional layer
model1.add(Conv2D(64, (3, 3), input_shape=INPUT_SHAPE))
model1.add(BatchNormalization())
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

# second convolutional layer
model1.add(Conv2D(128, (3, 3)))
model1.add(BatchNormalization())
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Flatten())
model1.add(Dense(256))
model1.add(BatchNormalization())
model1.add(Activation('relu'))
# Out layer
model1.add(Dense(NUM_LABELS))
model1.add(Activation('softmax'))

# summary model 1
model1.summary()
```

```python
# plot model 1
plot_model(model1, show_shapes=True, show_layer_names=True)
```

```python
# compile the model
model1.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# lets train our model for 20 epochs
res1 = model1.fit(train_generator,
                epochs = EPOCHS,
                verbose = 1,
                steps_per_epoch=STEP_SIZE_TRAIN,
                validation_steps=STEP_SIZE_TEST,
                validation_data = test_generator)  
```

```python
# plot the graph of accuracy
epochs_range = range(EPOCHS)

acc2 = res1.history['accuracy']
test_acc2 = res1.history['val_accuracy']
plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc2, label='Training Accuracy', color = "blue", marker = "o")
plt.plot(epochs_range, test_acc2, label='Test Accuracy', color = "red", marker = "o")
plt.legend(loc='lower right')
plt.title('Figure 9. CNN Adding Dropout Model: Training and Test Accuracy')
# plot the graph of loss
loss2 = res1.history['loss']
test_loss2 = res1.history['val_loss']
plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss2, label='Training Loss', color = "blue", marker = "o")
plt.plot(epochs_range, test_loss2, label='Test Loss', color = "red", marker = "o")
plt.legend(loc='upper right')
plt.title('Figure 10. CNN Adding Dropout Model: Training and Test Loss')
plt.show()
```

```python
# set up callback
es= tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', mode='max', patience=5, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=5, 
                                   verbose=1, mode='max', min_lr=0.00001)
callback=[es, reduce_lr]
```
